```python
import pandas as pd
import numpy as np

df= pd.read_csv('../DATA/cancer_classification.csv')
df.head()
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
X= df.drop('benign_0__mal_1', axis=1)
y=df['benign_0__mal_1']
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
from keras.callbacks import TensorBoard
board = TensorBoard(log_dir=loc,histogram_freq=1,write_graph=True, write_images=True,update_freq='epoch',embeddings_freq=1,profile_batch=2)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout

from keras.callbacks import EarlyStopping

early_stop= EarlyStopping(monitor='val_loss', patience=25, mode='min')

model = Sequential()
model.add(Dense(31, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(13, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=600, callbacks=[early_stop,board], validation_data=(X_train, y_train))
```
